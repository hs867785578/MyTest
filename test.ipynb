```python
def MyFmod(_X,  _Y):
    return _X - (int)(_X / _Y) * _Y

def NormalizeAngle(angle):	
    if angle> 2.0*pi or angle < (-2.0*pi):
        a = MyFmod(angle + pi, 2.0 * pi)
        if a < 0.0 :
            a += 2.0 * pi
        return a - 2.0*pi
    else:
        return angle
```
