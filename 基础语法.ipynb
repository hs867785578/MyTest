```python
#列表相加相当于append
a = [5,5]
b = [6]
c = [7]
d = [8]
print(a+b+c)


#如果想对应元素相加建议调用numpy
print(np.array(b)+np.array(c)+np.array(d))
```

```python
a = (5,5)
b = (6,)
c = (7,)
d = (8,)
print(a+b+c)
print(np.array(b)+np.array(c)+np.array(d))
```
