```python
from casadi import *
import time
```

```python
x = SX.sym('x')
y = SX.sym('y')
z = SX.sym('z')
nlp = {'x':vertcat(x,y,z), 'f':x**2+100*z**2, 'g':z+(1-x)**2-y}
S = nlpsol('S', 'ipopt', nlp)
print(S)
```

```python
start = time.perf_counter() 
r = S(x0=[2.5,3.0,0.75],lbg=0, ubg=0)
end =  time.perf_counter() 
runTime = end - start
```

```python
x_opt = r['x']
print('x_opt: ', x_opt)
print("运行时间：", runTime)
```
