```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#解决中午字体
plt.rcParams["font.sans-serif"]=["SimHei"]
plt.rcParams["font.family"]="sans-serif"

#解决符号无法显示
plt.rcParams["axes.unicode_minus"]=False
```

```python
x = np.linspace(-1,2,20)
y1 = np.e ** x -10
y2 = x ** 5 -4
plt.plot(x, y1, 'b--', label = '第一条线')
plt.plot(x, y2, 'r--', label = '第二条线')
plt.title('题目')
plt.xlabel('x坐标轴')
plt.ylabel('y坐标轴')
plt.legend()
plt.show()
```
